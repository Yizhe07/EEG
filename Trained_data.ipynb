```python
import pandas as pd
from keras.models import load_model
from sklearn.preprocessing import StandardScaler
import numpy as np

def preprocess_data(file_path):
    """
    Load and preprocess the EEG data from a given file path.
    """
    # Load the data from an Excel file
    data = pd.read_excel(file_path)
    
    # Select only EEG channel data
    X = data[['chn_1', 'chn_2', 'chn_3', 'chn_4', 'marker']].values
    
    # Instantiate and fit a StandardScaler
    scaler = StandardScaler().fit(X)
    # Transform the data using the fitted scaler
    X_scaled = scaler.transform(X)
    
    # Reshape data for the CNN (if your model requires it)
    X_reshaped = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)
    
    return X_reshaped

def predict_with_model(model_path, data_path):
    """
    Load the pre-trained model and predict on new data.
    """
    # Load the pre-trained model
    model = load_model(model_path)
    
    # Preprocess the new data
    X_new = preprocess_data(data_path)
    
    # Predict
    predictions = model.predict(X_new)
    predictions_classes = np.argmax(predictions, axis=1)
    
    # Interpret the predictions
    for prediction in predictions_classes:
        if prediction == 1:
            print("Sound Detected")
        else:
            print("No Sound Detected")

# Example usage
model_path = '/Users/Yizhe/Desktop/EEG_All_Data/Trained_Model.h5' 
data_path = '/Users/Yizhe/Desktop/EEG_All_Data/EEG_Data\Test_1.xlsx'   
predict_with_model(model_path, data_path)
```
